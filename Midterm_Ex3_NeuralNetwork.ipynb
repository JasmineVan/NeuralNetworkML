```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from tensorflow.keras import layers
from tensorflow.keras import models
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
```

# Classification

```python
col_names = ['Sex','Length','Diameter','Height','Whole weight','Shucked weight','Viscera weight','Shell weight','Rings']
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/ML/abalone.data", names=col_names)
df.head(5)
```

## Data preprocessing

```python
# Label encoder cho cột Sex
col_encode = ['Sex']

LE = LabelEncoder()
df[col_encode] = df[col_encode].apply(LE.fit_transform)
df.head()
```

```python
df['Rings'] = df['Rings'].apply(lambda x: 0 if x < 9 else (1 if x <= 10 else 2))
df.drop(df[df['Height']==0].index, axis=0)
```

```python
X = df.iloc[:,0:8]
print(X.shape)
print(X[:5])
```

```python
y = df.iloc[:,8]
print(y.shape)
print(y[:5])
```

```python
y = to_categorical(y)
print(y[:2])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

print('Train:')
print(X_train.shape)
print(y_train.shape)
print('Test:')
print(X_test.shape)
print(y_test.shape)
```

```python
model = models.Sequential()
model.add(layers.Dense(16, activation='tanh'))
model.add(layers.Dense(32, activation='sigmoid'))
model.add(layers.Dense(64, activation='sigmoid'))
model.add(layers.Dense(3,activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', 
              metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=50, batch_size=10, verbose=2)
```

```python
y_pred = model.predict(X_test)
print(y_pred[:2])
```

```python
y_pred = y_pred.argmax(axis=1)
y_pred = to_categorical(y_pred)
print(y_pred[:2])
print(y_test[:2])
print(y_pred.shape, y_test.shape)
```

```python
print(classification_report(y_test,y_pred))
```

# Regression

```python
df_reg = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/ML/hour.csv")
df_reg.head(5)
```

```python
df_reg.info()
```

```python
df_reg = df_reg.rename(columns={'weathersit':'weather',
                       'yr':'year',
                       'mnth':'month',
                       'hr':'hour',
                       'hum':'humidity',
                       'cnt':'count'})

df_reg = df_reg.drop(columns = ['instant' , 'dteday' , 'year'])
```

```python
df_reg.head(5)
```

```python
cols = ['weather','temp','atemp','humidity','windspeed','casual','registered']
df_reg[cols]
```

```python
X = df_reg[cols]
y = df_reg['count']
print(X[:2])
print(y[:2])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

print('Train:')
print(X_train.shape)
print(y_train.shape)
print('Test:')
print(X_test.shape)
print(y_test.shape)
```

```python
normalizer = layers.Normalization(axis=-1)
```

```python
normalizer.adapt(np.array(X_train))
```

```python
print(normalizer.mean.numpy())
```

```python
first = np.array(X[:1])

with np.printoptions(precision=2, suppress=True):
  print('First example:', first)
  print()
  print('Normalized:', normalizer(first).numpy())
```

```python
model = models.Sequential([
      normalizer,
      layers.Dense(64, activation='relu'),
      layers.Dense(64, activation='relu'),
      layers.Dense(1)     
])

model.compile(loss='mean_absolute_error', optimizer='adam', 
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(X_train,y_train,epochs=100, verbose=2)
```

```python
test_results = {}

test_results['dnn_model'] = model.evaluate(X_test, y_test, verbose=2)
```
